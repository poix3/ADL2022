```python
import matplotlib.pyplot as plt
import numpy as np

font1 = {'family' : 'Times New Roman',
'weight' : 'normal',
'size'   : 17,
}

origin_accuracy_1 = np.load('data/original_accuracy_1.npy')
origin_accuracy_2 = np.load('data/original_accuracy_2.npy')
origin_accuracy_3 = np.load('data/original_accuracy_3.npy')
batchNorm_accuracy_1 = np.load('data/batchNorm_accuracy_1.npy')
batchNorm_accuracy_2 = np.load('data/batchNorm_accuracy_2.npy')
batchNorm_accuracy_3 = np.load('data/batchNorm_accuracy_3.npy')
sgd_accuracy = np.load('data/SGD_accuracy.npy')

origin_loss_1 = np.load('data/original_loss_1.npy')
origin_loss_2 = np.load('data/original_loss_2.npy')
origin_loss_3 = np.load('data/original_loss_3.npy')
batchNorm_loss_1 = np.load('data/batchNorm_loss_1.npy')
batchNorm_loss_2 = np.load('data/batchNorm_loss_2.npy')
batchNorm_loss_3 = np.load('data/batchNorm_loss_3.npy')
sgd_loss = np.load('data/SGD_loss.npy')
```

```python
plt.figure(figsize=(15,7))

plt.subplot(2, 3, 1)
plt.plot(range(1,201),origin_accuracy_1)
plt.xlabel('epoch',font1)
plt.ylabel('accuray %',font1)
plt.title('Baseline test1', font1)
plt.subplot(2, 3, 2)
plt.plot(range(1,201),origin_accuracy_2)
plt.xlabel('epoch',font1)
plt.ylabel('accuray %',font1)
plt.title('Baseline test2', font1)
plt.subplot(2, 3, 3)
plt.plot(range(1,201),origin_accuracy_3)
plt.xlabel('epoch',font1)
plt.ylabel('accuray %',font1)
plt.title('Baseline test3', font1)

plt.subplot(2, 3, 4)
plt.plot(range(1,201),batchNorm_accuracy_1)
plt.xlabel('epoch',font1)
plt.ylabel('accuray %',font1)
plt.title('BatchNorm test1', font1)
plt.subplot(2, 3, 5)
plt.plot(range(1,201),batchNorm_accuracy_2)
plt.xlabel('epoch',font1)
plt.ylabel('accuray %',font1)
plt.title('BatchNorm test2', font1)
plt.subplot(2, 3, 6)
plt.plot(range(1,201),batchNorm_accuracy_3)
plt.xlabel('epoch',font1)
plt.ylabel('accuray %',font1)
plt.title('BatchNorm test3', font1)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,7))

plt.subplot(2, 3, 1)
plt.plot(range(1,201),origin_loss_1)
plt.xlabel('epoch',font1)
plt.ylabel('loss',font1)
plt.title('Baseline test1', font1)
plt.subplot(2, 3, 2)
plt.plot(range(1,201),origin_loss_2)
plt.xlabel('epoch',font1)
plt.ylabel('loss',font1)
plt.title('Baseline test2', font1)
plt.subplot(2, 3, 3)
plt.plot(range(1,201),origin_loss_3)
plt.xlabel('epoch',font1)
plt.ylabel('loss',font1)
plt.title('Baseline test3', font1)

plt.subplot(2, 3, 4)
plt.plot(range(1,201),batchNorm_loss_1)
plt.xlabel('epoch',font1)
plt.ylabel('loss',font1)
plt.title('BatchNorm test1', font1)
plt.subplot(2, 3, 5)
plt.plot(range(1,201),batchNorm_loss_2)
plt.xlabel('epoch',font1)
plt.ylabel('loss',font1)
plt.title('BatchNorm test2', font1)
plt.subplot(2, 3, 6)
plt.plot(range(1,201),batchNorm_loss_3)
plt.xlabel('epoch',font1)
plt.ylabel('loss',font1)
plt.title('BatchNorm test3', font1)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(range(1,201),sgd_accuracy)
plt.xlabel('epoch',font1)
plt.ylabel('accuracy',font1)
plt.title('SGD accuracy', font1)
plt.subplot(1,2,2)
plt.plot(range(1,201),sgd_loss)
plt.xlabel('epoch',font1)
plt.ylabel('loss',font1)
plt.title('SGD loss', font1)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,5))
plt.subplot(1, 2, 1)
l1, = plt.plot(range(1,201),(origin_accuracy_1+origin_accuracy_2+origin_accuracy_3)/3,c='red')
l2, = plt.plot(range(1,201),(batchNorm_accuracy_1+batchNorm_accuracy_2+batchNorm_accuracy_3)/3,c='blue')
l3, = plt.plot(range(1,201), sgd_accuracy, c='green')
plt.xlabel('epoch',font1)
plt.ylabel('accuray %',font1)
plt.legend(handles=[l1,l2,l3],labels=['Baseline','BatchNorm','SGD'],loc='best',prop=font1)
plt.title('Accuracy',fontdict=font1)
plt.subplot(1,2,2)
plt.subplot(1, 2, 2)
l1, = plt.plot(range(1,201),(origin_loss_1+origin_loss_2+origin_loss_3)/3,c='red')
l2, = plt.plot(range(1,201),(batchNorm_loss_1+batchNorm_loss_2+batchNorm_loss_3)/3,c='blue')
l3, = plt.plot(range(1,201), sgd_loss, c='green')
plt.xlabel('epoch',font1)
plt.ylabel('loss',font1)
plt.legend(handles=[l1,l2,l3],labels=['Baseline','BatchNorm','SGD'],loc='best',prop=font1)
plt.title('Loss',fontdict=font1)
plt.show()
```
