```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MaxNLocator

font1 = {'family' : 'Times New Roman',
'weight' : 'normal',
'size'   : 17,
}
```

```python
rotate = np.load('Rotation_losslist.npy')
gan_D = np.load('D_losslist.npy')
gan_G = np.load('G_losslist.npy')
autoencoder = np.load('Autoencoder_losslist.npy')
```

```python
plt.figure(figsize=(12,7))

plt.subplot(2,1,1)
l1, = plt.plot(range(1,31),rotate,c='red')
# l2, = plt.plot(range(1,151),gan_D+gan_G,c='blue')
l3, = plt.plot(range(1,151),gan_D,c='blue')
l4, = plt.plot(range(1,151),gan_G,c='orange')
l5, = plt.plot(range(1,21),autoencoder, c='green')
plt.xlabel('epoch',font1)
plt.ylabel('loss',font1)
plt.legend(handles=[l1,l3,l4,l5],labels=['Rotation','GAN:Discriminator','GAN:Generator','Autoencoder'],loc='best',prop=font1)
plt.title('Learning curves of pre-training',fontdict=font1)

plt.subplot(2,2,3)
plt.plot(range(1,31),rotate,c='red')
plt.xlabel('epoch',font1)
plt.ylabel('loss',font1)
plt.title('Rotation',fontdict=font1)

plt.subplot(2,2,4)
plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True))
plt.ylim(0,0.05)
plt.plot(range(1,21), autoencoder, c='green')
plt.xlabel('epoch',font1)
plt.ylabel('loss',font1)
plt.title('Autoencoder',fontdict=font1)

plt.tight_layout()
plt.show()
```
